```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from arcgis import GIS

gis = GIS()
item = gis.content.get("85d0ca4ea1ca4b9abf0c51b9bd34de2e")
flayer = item.layers[0]
df = flayer.query(where="NAME = 'New York' OR NAME = 'Los Angeles' OR NAME = 'Chicago' OR NAME = 'Houston' OR NAME = 'Philadelphia'").sdf

malePop = df["MALES"]
femPop = df["FEMALES"]
labels = df["NAME"]


x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, malePop, width, label='Male')
rects2 = ax.bar(x + width/2, femPop, width, label='Female')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Population')
ax.set_title('Male and Female City Population')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


fig.tight_layout()

plt.show()
```
